```python
import pandas as pd
import numpy as np
import time
from bs4 import BeautifulSoup
import requests
import urllib.request
```

```python
url = 'https://www.ilmeteo.it/portale/archivio-meteo/Pisa/2023/Gennaio'
r = requests.get(url)
soup = BeautifulSoup(r.content, "html.parser")
Tot = soup.find_all('div', {'class':'form-item'})
province = list()
for i in Tot:
    p = i.find_all('select',{'id':'edit-pid'}) 
    for t in p:
        n = t.find_all('option')
        for ss in n:
            if ss is not None:
                row = ss.text
                province.append('{} '.format(row))
```

```python
province
```

```python
# Province Cleaning            
P = list() 
for kk in province[1:]:
    kk=kk.strip()
    kk=kk.split()[0]
    P.append(kk)
print(P)
```

```python
Months = ['Gennaio','Febbraio','Marzo','Aprile','Maggio','Giugno','Luglio','Agosto','Settembre','Ottobre','Novembre','Dicembre']
```

```python
# subselection
PP =P[0:2]
```

```python
diz = dict()
for a in PP:
    time.sleep(5)
    print('Province {}:'.format(a))
    url = 'https://www.ilmeteo.it/portale/archivio-meteo/{}/2023/Gennaio'.format(a)
    req = requests.get(url)
    soup = BeautifulSoup(req.content, "html.parser")
    Tot = soup.find_all('div', {'class':'form-item'})
    comuni = list()
    for i in Tot:
        l = i.find_all('select',{'id':'edit-lid'})
        for m in l:
            h = m.find_all('option')
            for s in h:
                if s is not None:
                    row2 = s.text
                comuni.append('{} '.format(row2))
    C = list()
    for k in comuni[1:]:
        k=k.strip()
        C.append(k)
    for b in C:
        time.sleep(3)
        print(b)
        for j in Months[0:7]:
            url = "https://www.ilmeteo.it/portale/archivio-meteo/{}/2023/{}?format=csv".format(b,j)
            req = requests.get(url)
            url_content = req.content
            if len(url_content) < 5000: 
                csv_file = open('meteo2023.csv', 'ab')
                csv_file.write(url_content)
                csv_file.close()
                if b not in diz.keys():
                    diz[b] = a
            else:
                continue
```

```python
diz
```

```python
v = pd.read_csv('meteo2023.csv', sep=';')
v.drop_duplicates(inplace=True)
v
```

```python
v['PROVINCIA'] = v['LOCALITA'].map(diz)
v['FENOMENI'].fillna('Nessuno', inplace=True)
v.head(10)
```

```python
v.drop(v.index[0], inplace=True)
v.head()
```

```python
v.to_csv('meteo2023.csv',index=False)
```
